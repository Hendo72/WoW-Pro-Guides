```python
import pandas
import matplotlib.pyplot as plt
import matplotlib.dates as dates

%matplotlib inline
plt.rcParams['figure.figsize']=(20,10)
```

```python
data = pandas.read_csv("S3.csv", index_col=[0], parse_dates=True, date_format="%Y-%m-%d")
data = data.iloc[:,21:]
data.columns = map(lambda c: c[9:], data.columns)
```

```python
ax=data.plot(kind='area', title="WoWPro Downloads", figsize=(20,10))
ax.legend(loc='upper left', bbox_to_anchor=(0.0, 1.40),
          ncol=8, fancybox=True, shadow=True)
```

```python
x=data.loc["2014"].dropna(axis=1, how='all')
# x = x.iloc[: , :-2]
ax=x.plot(kind='area', title="WoWPro Downloads 2014", figsize=(20,10))
ax.legend(loc='upper left', bbox_to_anchor=(0.6, 1.05),
          ncol=4, fancybox=True, shadow=True)
```

```python
x = data.loc["2015"].dropna(axis=1, how='all')
x = x.iloc[: , 9:]
ax=x.plot(kind='area', title="WoWPro Downloads 2015", figsize=(20,10))
ax.legend(loc='upper left', bbox_to_anchor=(0.6, 1.05),
          ncol=4, fancybox=True, shadow=True)
#ax.set_ylim(0,1000)
x.sum(axis=0)
```

```python
x = data.loc["2016"].dropna(axis=1, how='all')
x = x.iloc[: , :-2]
ax=x.plot(kind='area', title="WoWPro Downloads 2016", figsize=(20,10))
ax.legend(loc='upper right', bbox_to_anchor=(0.5, 0.95),
          ncol=4, fancybox=True, shadow=True)
#ax.set_ylim(0,1000)
x.sum(axis=0)
```
